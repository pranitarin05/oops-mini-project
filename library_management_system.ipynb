```python
#library management system using oops concepts the system should calculate borrowing
#charges for different types of library items using oops concepts
# LibraryItem - Abstract class
# Book - child class
# Magazine - child class
# borrowing charge - encapsulated data
# LibraryApp - controller(HAS-A)
# output-
# Item Type: Book
# Borrow Days: 5
# Borrowing Charge: 50
# Item Type: Magazine
# Borrow Days: 3
# Borrowing Charge: 30

class LibraryItem:
    def __init__(self,item_id,title):
        self.item_id=item_id
        self.title=title
    def calculate_charge(self):
        pass

class Book(LibraryItem):
    def __init__(self,item_id,title,borrow_days):
        super().__init__(item_id,title)
        self._borrow_days=borrow_days
        #encapsulation because borrow days protected
    def calculate_charge(self):
        return self._borrow_days*10

class Magazine(LibraryItem):
    def __init__(self,item_id,title,borrow_days):
        super().__init__(item_id,title)
        self.borrow_days=borrow_days
    def calculate_charge(self):
        return self.borrow_days*10

class LibraryApp:
    def __init__(self):
        self.item=None
    def create_item(self,item_type):
        if item_type=="Book":
            self.item=Book(1,"Python Programming",5)
        else:
            self.item=Magazine(2,"Tech Magazine",3)
        return "Item Type: "+item_type
    def get_borrow_days(self):
        if hasattr(self.item,'_borrow_days'):
            return self.item._borrow_days
        else:
            return self.item.borrow_days
    def get_charge(self):
        return self.item.calculate_charge()

app=LibraryApp()
print(app.create_item("Book"))
print("Borrow Days:",app.get_borrow_days())
print("Borrowing Charge:",app.get_charge())
app2=LibraryApp()
print(app2.create_item("Magazine"))
print("Borrow Days:",app2.get_borrow_days())
print("Borrowing Charge:",app2.get_charge())
```
